```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
```

```python
# Load SBERT Model
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# Sample Resume & Job Description
resume_text = """
Python developer with experience in machine learning, Flask,
SQL databases, and data preprocessing.
"""

job_description = """
Looking for a Python developer with ML experience,
Flask framework knowledge, and SQL skills.
"""
```

```python
# Convert text to embeddings
resume_embedding = model.encode([resume_text])
jd_embedding = model.encode([job_description])
```

```python
score = cosine_similarity(resume_embedding, jd_embedding)[0][0]
print("Match Score:", round(score*100, 2), "%")
```

```python
skills = ['python', 'machine learning', 'flask', 'sql', 'docker']

resume_lower = resume_text.lower()
jd_lower = job_description.lower()

resume_skills = [s for s in skills if s in resume_lower]
jd_skills = [s for s in skills if s in jd_lower]

missing_skills = list(set(jd_skills) - set(resume_skills))

resume_skills, missing_skills
```
